## Imutável 🔒 vs Mutável 🔓

- Anteriormente vimos que **Strings são Imutáveis**.
- Você não pode mudar uma letra de uma string existente.

- Já as **Listas são Mutáveis**.
- Você pode alterar, remover e adicionar elementos livremente.

````python
# --- STRING (Imutável) ---
texto = "ProceXsamento"
print(f"Texto original: {texto}")

try:
    texto[5] = "s" # Tentar corrigir o 'X' para 's'
except TypeError as e:
    print(f"❌ Erro na String: {e}")

# Solução na String: Criar uma NOVA (Fatiamento + Concatenação)
texto_novo = texto[:5] + "s" + texto[6:]
print(f"Texto corrigido (nova string): {texto_novo}\n")

# --- LISTA (Mutável) ---
notas = [10, 5, 20, 9] # Ops, nota 20 não existe!
print(f"Lista original: {notas}")

notas[2] = 8 # Alteração direta funciona!
print(f"Lista corrigida: {notas}")
````

## 3. Manipulando Listas: O "CRUD" - create, read, update, delete

Uma lista é dinâmica. Precisamos saber adicionar, remover e combinar dados o tempo todo.

### 📥 Adicionar Elementos
* **`lista.append(item)`**: Adiciona o item no **final** da lista (o mais usado!).
* **`lista.insert(pos, item)`**: "Empurra" os elementos e insere o item em uma **posição específica**.

### 📤 Remover Elementos
* **`lista.pop()`**: Remove e retorna o **último** elemento (como tirar do topo de uma pilha).
* **`lista.remove(item)`**: Procura o **primeiro** item igual ao valor passado e o apaga. *Cuidado: Se não existir, dá erro!*

### ➕ Operadores Especiais
* **Soma (`+`)**: Junta duas listas em uma nova (Concatenação).
* **Multiplicação (`*`)**: Repete os elementos da lista N vezes (ótimo para inicializar vetores).

````python
mochila = ["Espada", "Escudo"]
print(f"Início: {mochila}")

mochila.append("Poção")
print(f"Append: {mochila}")

mochila.insert(0, "Mapa")
print(f"Insert: {mochila}")

item_usado = mochila.pop()
print(f"Pop:    {mochila} (Item retirado: {item_usado})")

if "Espada" in mochila:
    mochila.remove("Espada") # Joguei a espada fora
    print(f"Remove: {mochila}")

moedas = [10, 50, 100]
inventario_total = mochila + moedas
print(f"Soma (+):   {inventario_total}")

vidas = ["❤️"] * 3  #Win + .
print(f"Multipl (*): {vidas}")
````

## Percorrendo Listas (Iteração)

Existem duas formas principais de varrer uma lista com `for`:
1.  **Por Elemento (`for item in lista`):**
    - Mais Pythonico e limpo.
    - Use quando não precisa saber a posição.
2.  **Por Índice (`for i in range(len(lista))`):**
    - Use quando precisa alterar a lista original ou acessar posições matemáticas.

````python
valores = [11, 22, 33, 44]

print("Direto no Elemento (Leitura)")
for v in valores:
    print(f"Valor: {v}")

print("\nPelo Índice (Alteração)")
for i in range(len(valores)):
    valores[i] = valores[i] * 2
    print(f"Índice {i} agora tem {valores[i]}")

print(f"\nLista Final: {valores}")
````

##  Algoritmos Clássicos: Implementação Manual

- Embora o Python ofereça funções como `min()`, `remove()` e o operador `==`,
- É essencial compreender a lógica algorítmica delas.

Vamos ver 3 exemplos:

* **Filtragem:** Criar uma nova lista excluindo um elemento específico (diferente do `.remove()` que apaga apenas o primeiro).

* **Busca Linear:** Encontrar o menor valor percorrendo a lista.

* **Comparação:** Verificar se duas listas são idênticas elemento a elemento.

````python
def remover_todas_ocorrencias(valor, lista_original: list) -> list:
    nova_lista = []

    for item in lista_original:
        if item != valor:
            nova_lista.append(item)

    return nova_lista

dados = [11, 33, 55, 33, 66, 33]
limpos = remover_todas_ocorrencias(33, dados)

print(f"Original: {dados}")
print(f"Sem o 33: {limpos}")
````

````python
def encontrar_menor(lista: list) -> int:
    menor = lista[0]
    for i in range(1, len(lista)):
        if lista[i] < menor:
            menor = lista[i]
    return menor

valores = [11, 22, 33, 44, -11, 77, 55, 99]
resultado = encontrar_menor(valores)

print(f"Lista: {valores}")
print(f"Menor valor encontrado: {resultado}")
````

````python
def sao_iguais(lista_a: list, lista_b: list) -> bool:
    if len(lista_a) != len(lista_b):
        return False

    for i in range(len(lista_a)):
        if lista_a[i] != lista_b[i]:
            return False

    return True

v1 = [8, -9, 1, 0, 0]
v2 = [8, -9, 1, 0, 3]

print(f"Comparando v1 e v2: {sao_iguais(v1, v2)}")
print(f"Comparando v1 e v1: {sao_iguais(v1, v1)}")
````

## Buscando Dados (`in`, `count`, `index`) 🔎


- O Python tem operadores especiais para encontrar um item na lista e saber onde ele está.
- Funciona  como perguntas em inglês.

* **`item in lista`**: Retorna `True` se o item existir (Pergunta: *"Está dentro?"*).
* **`lista.count(item)`**: Conta quantas vezes o item aparece.
* **`lista.index(item)`**: Diz a posição (índice) da **primeira** vez que o item aparece.

````python
convidados = ["Ana", "Carlos", "Beatriz", "Ana", "Daniel"]

busca = "Ana"

if busca in convidados:
    print(f"✅ {busca} está na lista!")
else:
    print(f"❌ {busca} não foi convidado(a).")

qtd = convidados.count(busca)
print(f"O nome {busca} aparece {qtd} vezes.")

if busca in convidados:
    posicao = convidados.index(busca)
    print(f"A primeira {busca} está na posição {posicao}.")
````

### 🧠 Cálculo da Mediana

- A mediana é o valor que está no **meio** da lista **ordenada** (vamos precisar ordenar a lista).
  * Se a lista tem tamanho ímpar: é o elemento central.
  * Se tem tamanho par: é a média dos dois centrais.



````python
def calcular_mediana(lista):
    lista.sort()
    n = len(lista)

    if n % 2 == 1:
        return lista[n // 2]
    else:
        meio1 = lista[n // 2 - 1]
        meio2 = lista[n // 2]
        return (meio1 + meio2) / 2

print(f"Mediana [1, 3, 5]: {calcular_mediana([5, 1, 3])}")
print(f"Mediana [1, 2, 3, 4]: {calcular_mediana([4, 1, 3, 2])}")
````

### 🧠  Concatenar (`+`) e `extend()`

- Existem  duas formas de juntar listas.
- Elas parecem iguais, mas agem diferente na memória!

1.  **`lista1 + lista2`**: Cria uma **NOVA** lista (gasta mais memória).
    - A lista original fica intacta.
2.  **`lista1.extend(lista2)`**: Modifica a **lista1** original
    - Adicionando os itens na lista.
    - É mais eficiente.

````python
lista_a = [1, 2]
lista_b = [3, 4]
nova_lista = lista_a + lista_b

print(f"Com Soma (+): {nova_lista}")
print(f"A lista 'a' mudou? {lista_a} (Não!)")

lista_a = [1, 2]
lista_a.extend(lista_b)

print(f"\nCom Extend: {lista_a}")
print("A lista 'a' foi alterada diretamente!")
````

## Lendo Listas do Usuário (`input`)

- Existem duas formas principais de preencher uma lista com dados do usuário.

### 📝 Método 1: Item por Item (Loop)
- Este é o método clássico.
- Usamos quando queremos controlar exatamente quantos itens entrarão.
  1. Perguntamos a **quantidade** (`n`).
  2. Criamos uma lista vazia.
  3. Fazemos um laço `for` rodar `n` vezes, fazendo `append` a cada volta.

````python
n = int(input("Quantos números você vai digitar? "))

lista_usuario = []

for i in range(n):
    numero = int(input(f"Digite o {i+1}º valor: "))
    lista_usuario.append(numero)

print(f"Sua lista final: {lista_usuario}")
````

### 📝 Método 2: Lendo Listas do Usuário (usando `split`)

- O `input()` lê tudo como uma string só.
- Usamos o método `.split()` para \"fatiar\" a string nos espaços em branco.

````python
texto_usuario = "10 20 30 40"

lista_strings = texto_usuario.split()
print(f"Split puro: {lista_strings}")

lista_numeros = []
for item in lista_strings:
    lista_numeros.append(int(item))

print(f"Convertido: {lista_numeros} | Tipo do 1º item: {type(lista_numeros[0])}")
````

## Sintaxe Concisa e List Comprehension

- Python oferece formas de escrever códigos mais compactos.
- Embora pareçam complexos inicialmente, são padrões da indústria e frequentemente utilizados em sugestões de IAs.

### Expressões Condicionais (`if-else` em uma linha)
- Em muitas linguagens, existe o "operador ternário".
- Em Python, podemos atribuir valores baseados em uma condição numa única linha.

* **Sintaxe:** `valor_se_verdadeiro if condicao else valor_se_falso`

### Compreensão de Lista (List Comprehension)
- Esta é a forma "Pythonica" de criar listas baseadas em outras iteráveis (listas, ranges, strings).
- Ela substitui o padrão clássico de: *criar lista vazia -> laço for -> append*.

* **Sintaxe Básica:** `[expressao for item in lista]`
* **Com Filtro:** `[expressao for item in lista if condicao]`

````python
nota = 8
situacao = "Aprovado" if nota >= 5 else "Reprovado"
print(f"Situação: {situacao}")

print("\n--- Gerando Quadrados ---")

quadrados_classico = []
for i in range(10):
    quadrados_classico.append(i**2)

quadrados_pro = [i**2 for i in range(10)]

print(f"Clássico: {quadrados_classico}")
print(f"Conciso:  {quadrados_pro}")

print("\n--- Classificando Números ---")
numeros = [1, 2, 3, 4, 5]

paridade = ['Par' if n % 2 == 0 else 'Ímpar' for n in numeros]
print(f"Rótulos: {paridade}")

apenas_pares = [n for n in numeros if n % 2 == 0]
print(f"Filtrados: {apenas_pares}")
````

### ⚡ Aplicação Prática: Leitura de Dados (Moodle/VPL)

- Em exercícios de programação competitiva ou no Moodle, é comum recebermos vários dados na **mesma linha**, separados por espaço.
* **Entrada:** `10 20 30 40 50`
* **Objetivo:** Transformar isso na lista de inteiros `[10, 20, 30, 40, 50]`

Usando *List Comprehension*, resolvemos isso em uma única linha, combinando `input()`, `split()` e conversão de tipo.

````python
entrada_usuario = "10.5 20.0 30.2 40.8"

# Passo a passo do que o Python faz nesta linha:
# 1. input(): Lê a string "10.5 20.0..."
# 2. .split(): Quebra em lista de strings ['10.5', '20.0'...]
# 3. for x in...: Percorre cada string
# 4. float(x): Converte cada pedaço para número real
vetor = [float(x) for x in entrada_usuario.split()]

print(f"Lista processada: {vetor}")
print(f"Tipo do primeiro elemento: {type(vetor[0])}")

# Exemplo: Filtrando dados na leitura
# Lê apenas se o número for maior que 20
maiores_que_20 = [float(x) for x in entrada_usuario.split() if float(x) > 20]
print(f"Filtro na entrada (>20): {maiores_que_20}")
````

### 🧠 Exercício Remoção Completa

- O método `.remove()` do Python apaga apenas a **primeira** ocorrência de um valor.
- Para remover **todas** as ocorrências de um valor (ex: remover todos os zeros de um vetor), a *List Comprehension* é a ferramenta ideal, pois cria uma nova lista contendo apenas o que queremos manter.

````python
def remover_todas_ocorrencias(lista_original, valor_proibido):
    return [item for item in lista_original if item != valor_proibido]

dados = [1, 0, 2, 0, 3, 0, 4]
limpos = remover_todas_ocorrencias(dados, 0)

print(f"Original: {dados}")
print(f"Sem zeros: {limpos}")
````

## 🔬 Parte 2: Sob o Capô (Memória e Performance)

- Até agora usamos listas pela sua facilidade.
- Mas como o computador guarda a lista  na memória RAM?

* **Variáveis:** Ocupam espaço físico (Bytes) na memória.
* **Listas Python:** Como aceitam mistura de tipos (`[1, "a", True]`),
    - Elas não guardam os dados juntos.
    - Elas guardam **Endereços** (Ponteiros) para onde os dados estão espalhados na memória.
    - **Custo:** Isso gasta mais memória e é mais lento para processar.

Vamos usar o módulo `sys` para ver o peso real das variáveis.

### ⚠️ Referência (Ponteiros)

- Como listas podem ser gigantes, o Python tenta economizar memória não criando cópias automaticamente.
- * Se você fizer `A = B`, você **não** criou uma cópia.
- Você criou um **Apelido** para a mesma lista na memória.
- Se mexer em `A`, **estraga** `B`.

````python
lista_original = [1, 2, 3]
lista_fake = lista_original

lista_fake[0] = 999
print(f"Original foi alterada?! {lista_original}")

lista_original = [1, 2, 3]
lista_real = lista_original.copy()

lista_real[0] = 999
print(f"Original preservada: {lista_original}")
print(f"Cópia alterada:      {lista_real}")
````

## 🚀 Arrays de Verdade (`array` e `numpy`)

- Para processamento pesado (Engenharia, Data Science, Imagens), as Listas do Python são ineficientes.
- A solução é usar **Arrays Tipados**.

* **Característica:** Exigem que todos os dados sejam do **mesmo tipo** (só inteiros, ou só floats).
* **Vantagem:** São armazenados de forma contígua (coladinhos) na memória. Muito mais leves e rápidos.

Temos o módulo nativo `array` (simples) e o poderoso `numpy` (padrão de mercado).

````python
import sys
import array
import numpy as np

def bytes_to_mb(bytes_val):
    return bytes_val / (1024 * 1024)

QTD = 1_000_000
lista_py = [0] * QTD
array_py = array.array('i', lista_py)
array_np = np.array(lista_py, dtype='int32')

print(f"--- Comparando Memória para {QTD} inteiros ---")

mem_lista = sys.getsizeof(lista_py)
mem_array = sys.getsizeof(array_py)
mem_numpy = array_np.nbytes

print(f"Python List: {bytes_to_mb(mem_lista):.2f} MB")
print(f"Array Nativo: {bytes_to_mb(mem_array):.2f} MB")
print(f"Numpy Array:  {bytes_to_mb(mem_numpy):.2f} MB")

print("\nConclusão: Arrays ocupam muito menos espaço!")
````

## ⏱️ Medindo Performance (Tempo)

- Além de memória, Arrays são projetados para velocidade.
- Vamos usar o módulo `time` para cronometrar quanto tempo leva para somar 10 milhões de números.

````python
import time

QTD = 10_000_000
lista_grande = list(range(QTD))
array_grande = np.array(lista_grande)

print(f"🏁 Disputa: Somando {QTD} números...")

inicio = time.time()
soma = 0
for x in lista_grande:
    soma += x
fim = time.time()
print(f"🐢 Loop Manual: {fim - inicio:.4f} seg")

inicio = time.time()
soma = np.sum(array_grande)
fim = time.time()
print(f"🐇 Numpy.sum:  {fim - inicio:.4f} seg")
````

## 🤖 Aplicação Prática: Modelando um Neurônio Artificial com Vetores

A base de toda Rede Neural Artificial moderna, desde perceptrons simples até Grandes Modelos de Linguagem (LLMs), fundamenta-se em operações de Álgebra Linear.

Um neurônio artificial opera realizando uma **combinação linear** de suas entradas.

Estruturalmente, um neurônio é composto por:
1.  **Entradas ($x$):** Vetor de dados provenientes de sensores ou de outros neurônios.
2.  **Pesos ($w$):** Vetor de coeficientes que ponderam a importância de cada entrada (o "conhecimento" da rede).
3.  **Processamento (Soma Ponderada):** Ocorre o somatório do produto de cada entrada pelo seu peso.

Matematicamente (e simplificando), definimos o **Potencial de Ativação** ($u$) como:
$$u = \sum_{i=0}^{n} (x_i \cdot w_i) + b$$

### 🚀 Por que usar Arrays (Numpy) e não Listas?
- Em Python puro, calcular esse somatório exigiria um laço `for` para iterar item a item, o que é computacionalmente custoso para milhões de dados.
- Ao utilizarmos **Arrays** da biblioteca `numpy`, podemos realizar a operação de **Produto Escalar (Dot Product)**.
- Isso processa todos os números simultaneamente (vetorização), sendo ordens de grandeza mais rápido e semanticamente mais claro.

---

### ☔ Estudo de Caso: Sistema de Decisão Pluviométrica
Vamos implementar um neurônio classificador binário que determina a ação: **"Levar guarda-chuva?"**.

#### 1. Definição do Vetor de Entradas (Sensores)
Para o computador, a informação dos sensores deve ser convertida em números e organizada em estruturas que chamamos de **Vetores** (ou tecnicamente, *Tensores 1D*).

* $x_0$ **(Estado do Céu):** `1` (Nublado/Sim)
* $x_1$ **(Previsão Meteorológica):** `0.8` (80% de probabilidade de chuva)


#### 2. Definição do Vetor de Pesos (Conhecimento)
Os pesos representam a confiabilidade ou influência de cada sensor na decisão final.
* $w_0$ **(Peso Visual):** `2.0` (Baixa confiabilidade; nuvens enganam).
* $w_1$ **(Peso do App):** `10.0` (Alta confiabilidade; baseada em dados).

#### 🧮 Processamento Vetorial
O neurônio calculará o risco total através do produto escalar:
$$\text{Risco} = \vec{x} \cdot \vec{w} = (1 \times 2.0) + (0.8 \times 10.0) = 10.0$$

**Função de Ativação (Limiar):** Se o Risco calculado for superior a **5.0**, o neurônio \"dispara\" (ativação positiva).

````python
import numpy as np

def neuronio_vetorial(entradas: np.array, pesos: np.array) -> float:
    """
    Simula um neurônio artificial utilizando operações vetoriais (Produto Escalar).
    A ausência de laços 'for' explícitos torna esta função otimizada para alto desempenho.
    """
    # O método .dot realiza a soma dos produtos (x1*w1 + x2*w2 ...)
    potencial_ativacao = np.dot(entradas, pesos)

    return potencial_ativacao

# --- 1. Inicialização dos Vetores ---
# Entradas: [Céu Nublado (bool), Probabilidade Chuva (float)]
sensores = np.array([1, 0.8])

# Pesos: [Importância Visual, Importância do App]
conhecimento = np.array([2.0, 10.0])

print(f"📡 Sensores (Input): {sensores}")
print(f"⚖️ Pesos (Weights):  {conhecimento}")

# --- 2. Processamento (Feedforward) ---
# Cálculo direto sem loop for (Vetorização)
risco_calculado = neuronio_vetorial(sensores, conhecimento)

print(f"\n⚡ Potencial de Ativação (Risco): {risco_calculado:.2f}")

# --- 3. Função de Ativação (Decisão) ---
limiar = 5.0

if risco_calculado > limiar:
    print(f"📢 Decisão: {risco_calculado} > {limiar} -> ☂️ ATIVAR: Levar guarda-chuva.")
else:
    print(f"📢 Decisão: {risco_calculado} <= {limiar} -> 😎 REPOUSO: Não levar nada.")
````
