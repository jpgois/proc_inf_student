## 🍕 Fatiamento (Slicing): Extraindo Pedaços de Listas e Strings

- Muitas vezes, não queremos acessar apenas um elemento, mas sim um subconjunto inteiro (uma "fatia") da nossa lista ou string.
- O Python possui uma sintaxe para isso: `lista[inicio:fim:passo]`.

* **`inicio`**: Onde a fatia começa (inclusivo).
* **`fim`**: Onde a fatia termina (exclusivo, ou seja, para no elemento anterior).
* **`passo`**: O tamanho do salto entre os elementos.


Se omitirmos o `inicio`, o Python assume o índice 0. Se omitirmos o `fim`, ele vai até o último elemento.

````python
### GABARITO ###
frutas = ['maçã', 'banana', 'cereja', 'damasco', 'figo']
print(f"Lista original: {frutas}")

# Pegando do índice 1 ao 2 (o 3 é exclusivo)
print(f"Fatia [1:3]: {frutas[1:3]}")

# Pegando do início até o final, pulando de 2 em 2
print(f"Fatia [0:5:2]: {frutas[0:5:2]}")

# O fatiamento funciona EXATAMENTE igual para Strings!
texto = "Processamento"
print(f"\nString original: {texto}")
print(f"Fatia [0:8]: {texto[0:8]}")
print(f"Fatia invertida [::-1]: {texto[::-1]}") # Truque clássico para inverter!
````

### 🧠 Cálculo da Mediana

A mediana é o valor que está exatamente no **meio** da lista. Mas atenção: o computador não sabe quem é o meio se os dados estiverem bagunçados. **A lista precisa estar ordenada** primeiro!

* **Se a lista tem tamanho ÍMPAR:** A mediana é o próprio elemento central.
* **Se a lista tem tamanho PAR:** Não existe um único "meio". A mediana será a média aritmética dos dois elementos centrais.

#### 📊 Visualizando o Processo

**1️⃣ Caso Ímpar (Ex: 5 elementos)**
Primeiro recebemos os dados desordenados e organizamos do menor para o maior. O elemento que sobrar isolado no centro é a nossa resposta.

```text
Vetor Desordenado: [ 9,  2, 15,  7,  4 ]

Vetor Ordenado:    [ 2,  4,  7,  9, 15 ]
                             ⬆
                         Mediana = 7
```
**2️⃣ Caso Par (Ex: 6 elementos)**
Aqui, após ordenar, não há um centro único. Pegamos os dois valores que dividem o vetor ao meio e calculamos a média entre eles.

```text
Vetor Desordenado: [ 20,  4,  9,  2, 15,  7 ]

Vetor Ordenado:    [  2,  4,  7,  9, 15, 20 ]
                              ⬆   ⬆
                              |   |
                    Mediana = (7 + 9) / 2 = 8
```                    

## 🧮 Produto Escalar (Dot Product)

Dado um vetor $\vec{u} = (u_1, u_2, \dots, u_n)$ e um vetor $\vec{v} = (v_1, v_2, \dots, v_n)$, o produto escalar é a soma da multiplicação elemento a elemento:
$$\vec{u} \cdot \vec{v} = \sum_{i=0}^{n-1} u_i \cdot v_i$$



````python
### GABARITO ###
def produto_escalar_manual(vetor_u, vetor_v):
    if len(vetor_u) != len(vetor_v):
        return "Erro: Os vetores precisam ter o mesmo tamanho!"

    soma_total = 0
    for i in range(len(vetor_u)):
        soma_total += vetor_u[i] * vetor_v[i]

    return soma_total

# Vetores de teste
u = [1, 0.8, 0.5]
v = [2.0, 10.0, 5.0]

resultado_manual = produto_escalar_manual(u, v)
print(f"Vetor U: {u}")
print(f"Vetor V: {v}")
print(f"Produto Escalar Calculado Manualmente: {resultado_manual}")
````

## 🔎 Busca Linear (Sequencial)

- A forma mais intuitiva de procurar uma informação em uma lista é olhar elemento por elemento, do início ao fim, até encontrar o que queremos (ou chegar ao final e descobrir que não está lá).
- Esse processo é chamado de **Busca Linear**.

- Embora o Python tenha o operador `in` (ex: `if alvo in lista:`), entender como a busca linear funciona é essencial.
- É um algoritmo simples, mas que pode ser demorado se a lista for gigantesca, pois no pior dos casos ele precisará verificar todos os elementos.

````python
### GABARITO ###
def busca_linear(lista, alvo):
    """
    Procura um 'alvo' na 'lista'.
    Retorna o índice onde foi encontrado ou -1 se não existir.
    """
    for i in range(len(lista)):
        if lista[i] == alvo:
            return i  # Encontrou! Retorna a posição.

    return -1  # Percorreu tudo e não achou.

dados = [15, 8, 23, 42, 4, 16]
valor_procurado = 42

posicao = busca_linear(dados, valor_procurado)

if posicao != -1:
    print(f"Busca Linear: Valor {valor_procurado} encontrado no índice {posicao}.")
else:
    print(f"Busca Linear: Valor {valor_procurado} não existe na lista.")
````

## 🚀 Busca Binária (A estratégia de "Dividir para Conquistar")

- Imagine procurar uma palavra em um dicionário físico de 1000 páginas.
- Você não lê página por página (Busca Linear).
- Você abre no meio: se a palavra que você quer vem antes, você ignora a metade final e foca só na primeira metade.
- E repete o processo.

- Essa é a **Busca Binária**.
- Ela é incrivelmente mais rápida que a busca linear, mas tem uma condição necessária: **A lista DEVE estar ordenada**.

````python
### GABARITO ###
def busca_binaria(lista_ordenada, alvo):
    """
    Procura um 'alvo' dividindo a lista pela metade sucessivamente.
    Retorna o índice ou -1 se não encontrar.
    """
    esq = 0
    dir = len(lista_ordenada) - 1

    while esq <= dir:
        meio = (esq + dir) // 2  # Pega o índice central (divisão inteira)

        if lista_ordenada[meio] == alvo:
            return meio          # Encontrou o alvo no meio!

        elif lista_ordenada[meio] < alvo:
            esq = meio + 1       # O alvo é maior, deve estar na metade da direita

        else:
            dir = meio - 1       # O alvo é menor, deve estar na metade da esquerda

    return -1 # Se os ponteiros se cruzarem, o alvo não está na lista

# A lista OBRIGATORIAMENTE precisa estar ordenada!
dados_ordenados = [4, 8, 15, 16, 23, 42]
valor_procurado = 16

posicao = busca_binaria(dados_ordenados, valor_procurado)

if posicao != -1:
    print(f"Busca Binária: Valor {valor_procurado} encontrado no índice {posicao}.")
else:
    print(f"Busca Binária: Valor {valor_procurado} não existe na lista.")
````

- meio = (3 + 5) // 2 = 4

- esq = 3, dir = 3
- meio = (3 + 3) // 2 = 3

- 16 == 16? SIM!
- Retorna o índice 3 e encerra a função.

## 🃏 Algoritmos Clássicos: Ordenação por Inserção (Insertion Sort)

- O Python facilita a nossa vida com o método `.sort()`, que ordena a lista automaticamente e a altera diretamente na memória.
- Mas como o computador faz isso nos bastidores?

-Existem muitos algoritmos de ordenação.
- Vamos ver o   **Insertion Sort**, simula a forma como organizamos cartas de baralho na mão:
1. Pegamos uma carta de cada vez.
2. Comparamos com as cartas que já estão na nossa mão (à esquerda).
3. Inserimos a carta na sua posição correta.

### 🔍 Teste de Mesa: Insertion Sort Passo a Passo

Vamos acompanhar o que acontece com a lista `[87, 31, 546, 97, -15, 654]` a cada iteração do nosso laço `for`.
A barra vertical `|` separa a parte da lista que já está ordenada (à esquerda) da parte que ainda será processada (à direita).

**Estado Inicial:** `[87 | 31, 546, 97, -15, 654]`
*(O primeiro elemento, 87, já é considerado nossa mão inicial ordenada).*

* **Passo 1 (i=1):** A `chave` é **31**.
  * Comparamos com 87. Como 87 > 31, movemos o 87 para a frente e inserimos o 31 na primeira posição.
  * **Lista atual:** `[31, 87 | 546, 97, -15, 654]`

* **Passo 2 (i=2):** A `chave` é **546**.
  * Comparamos com 87. Como 87 não é maior que 546, o elemento fica onde está.
  * **Lista atual:** `[31, 87, 546 | 97, -15, 654]`

* **Passo 3 (i=3):** A `chave` é **97**.
  * Comparamos com 546 (maior, move para frente) e depois com 87 (menor, paramos). Inserimos o 97.
  * **Lista atual:** `[31, 87, 97, 546 | -15, 654]`

* **Passo 4 (i=4):** A `chave` é **-15**.
  * Comparamos com 546, 97, 87 e 31. Todos são maiores e se movem para a frente. O -15 vai parar no início da lista.
  * **Lista atual:** `[-15, 31, 87, 97, 546 | 654]`

* **Passo 5 (i=5):** A `chave` é **654**.
  * Comparamos com 546. Como não é maior, o 654 fica onde está. Fim do laço!
  * **Lista final:** `[-15, 31, 87, 97, 546, 654]`

````python
### GABARITO ###
def insertion_sort(lista):
    # Começamos do segundo elemento (índice 1), pois o primeiro já "está na mão"
    for i in range(1, len(lista)):
        chave = lista[i]
        j = i - 1

        # Move os elementos da lista que são maiores que a chave
        # para uma posição à frente de sua posição atual
        while j >= 0 and lista[j] > chave:
            lista[j + 1] = lista[j]
            j -= 1

        lista[j + 1] = chave
        # Print opcional para ver o passo a passo:
        # print(f"Passo {i}: {lista}")

dados_desordenados = [87, 31, 546, 97, -15, 654]
print(f"Antes da ordenação: {dados_desordenados}")

insertion_sort(dados_desordenados)
print(f"Após o Insertion Sort: {dados_desordenados}")
````

## 🔬 Parte 2: Sob o Capô (Memória e Performance)

- Até agora usamos listas pela sua facilidade.
- Mas como o computador guarda a lista  na memória RAM?

* **Variáveis:** Ocupam espaço físico (Bytes) na memória.
* **Listas Python:** Como aceitam mistura de tipos (`[1, "a", True]`),
    - Elas não guardam os dados juntos.
    - Elas guardam **Endereços** (Ponteiros) para onde os dados estão espalhados na memória.
    - **Custo:** Isso consome mais memória e é mais lento para processar.

## 🚀 Arrays de Verdade (`array` e `numpy`)

- Para processamento pesado (Engenharia, Data Science, Imagens), as Listas do Python são ineficientes.
- A solução é usar **Arrays Tipados**.

* **Característica:** Exigem que todos os dados sejam do **mesmo tipo** (só inteiros, ou só floats).
* **Vantagem:** São armazenados de forma contígua (lado a lado) na memória.
* Muito mais leves e rápidos.

Temos o módulo nativo `array` (simples) e o mais sofisticado `numpy` (padrão de mercado).

### 🛠️ Sintaxe Básica

Antes de testarmos o limite da memória, como criamos essas estruturas na prática?

A regra geral é simples: normalmente, nós pegamos uma lista comum do Python (ou geramos números) e "convertemos" para um Array tipado.

1. **`array`:** Exige que você passe um "código de tipo" (uma letra) para avisar o computador qual será o tamanho da caixa de memória. Exemplo: `'i'` para Inteiros, `'d'` para Decimais (Double).
2. **`numpy`:** É muito mais inteligente e amigável. Você passa o `dtype` (data type) ou, muitas vezes, ele mesmo adivinha o tipo automaticamente.

````python
import sys
import array
import numpy as np

def bytes_to_mb(bytes_val):
    return bytes_val / (1024 * 1024)

QTD = 1_000_000
lista_py = [0] * QTD
array_py = array.array('i', lista_py)
array_np = np.array(lista_py, dtype='int32')

print(f"--- Comparando Memória para {QTD} inteiros ---")

mem_lista = sys.getsizeof(lista_py)
mem_array = sys.getsizeof(array_py)
mem_numpy = array_np.nbytes

print(f"Python List: {bytes_to_mb(mem_lista):.2f} MB")
print(f"Array Nativo: {bytes_to_mb(mem_array):.2f} MB")
print(f"Numpy Array:  {bytes_to_mb(mem_numpy):.2f} MB")

print("\nConclusão: Estruturas do tipo Array (nativo ou NumPy) são muito mais eficientes em memória do que as Listas nativas do Python.")
````

### ⏱️ Medindo Performance (A Disputa Justa)

Além de economizar memória, os Arrays do NumPy são projetados para **velocidade**.
Mas para sermos justos, não podemos comparar o NumPy apenas com um `for` loop manual (que é naturalmente lento). O Python possui sua própria função otimizada de soma, o `sum()`.

Vamos usar o módulo `time` para cronometrar e fazer uma disputa tripla somando **10 milhões de números**:

1. 🐢 **Loop Manual:** O jeito mais lento, pois o Python lê linha por linha.
2. 🐕 **Função `sum()` nativa:** O jeito inteligente do Python. Ela é escrita na linguagem C (muito rápida), mas ainda precisa lidar com o peso das Listas flexíveis.
3. 🚀 **Função `np.sum()` do NumPy:** O topo da performance. Escrita em C e otimizada matematicamente para somar blocos inteiros de memória de uma só vez (Vetorização).

````python
import time
import numpy as np

QTD = 10_000_000
lista_grande = list(range(QTD))
array_grande = np.array(lista_grande, dtype='int32')

print(f"🏁 Disputa: Somando {QTD} números...\n")

# --- Competidor 1: Loop Manual ---
inicio = time.time()
soma = 0
for x in lista_grande:
    soma += x
fim = time.time()
print(f"🐢 1. Loop Manual (Python puro):  {fim - inicio:.4f} segundos")

# --- Competidor 2: Função sum() do Python ---
inicio = time.time()
soma = sum(lista_grande)
fim = time.time()
print(f"🐕 2. Função sum() nativa (Lista): {fim - inicio:.4f} segundos")

# --- Competidor 3: np.sum() do NumPy ---
inicio = time.time()
soma = np.sum(array_grande)
fim = time.time()
print(f"🚀 3. Função np.sum() (NumPy):     {fim - inicio:.4f} segundos")

print("-" * 45)
print("💡 Conclusão: O sum() nativo é excelente e esmaga o loop manual.")
print("Mas o NumPy ainda vence por lidar com Arrays puros e usar vetorização no processador!")
````

## 🤖 Aplicação Prática: Modelando um Neurônio Artificial com Vetores

A base de toda Rede Neural Artificial moderna, desde perceptrons simples até Grandes Modelos de Linguagem (LLMs), fundamenta-se em operações de Álgebra Linear.

</div>

Um neurônio artificial opera realizando uma **combinação linear** de suas entradas. Antes de olharmos para a rede inteira acima, vamos entender a anatomia de um **único neurônio**:

#### Diagrama de um Único Neurônio Artificial

```text
   Entradas (x)      Pesos (w)        Processamento (Neurônio)
   ------------      ---------        -----------------------

   ( x0 ) ---------> [ w0 ] -----\
                                  \       Soma Ponderada
   ( x1 ) ---------> [ w1 ] ----------> Σ(xi * wi) + b ---> Saída
                                  /
   ( .. ) ---------> [ .. ] -----/
                                /
   ( bias ) -------------------/
     (b)
```


Estruturalmente, um neurônio é composto por:
1.  **Entradas ($x$):** Vetor de dados provenientes de sensores ou de outros neurônios.
2.  **Pesos ($w$):** Vetor de coeficientes que ponderam a importância de cada entrada (o "conhecimento" da rede).
3.  **Processamento (Soma Ponderada):** Ocorre o somatório do produto de cada entrada pelo seu peso correspondente.

Matematicamente (e simplificando), definimos o **Potencial de Ativação** ($u$) como:
$$u = \sum_{i=0}^{n} (x_i \cdot w_i) + b$$

### 🚀 Por que usar Arrays (NumPy) e não Listas?
* Se tentarmos multiplicar duas listas comuns (`vetor_x * vetor_w`), o Python retornará um **erro**. Listas nativas não sabem fazer diretamente operações matemáticas entre si.
* Em Python puro, a única saída seria usar um laço `for` para iterar, parear e multiplicar item a item, o que é computacionalmente lento para milhões de dados.
* Ao utilizarmos **Arrays** da biblioteca `numpy`, habilitamos a matemática vetorial. Podemos realizar a operação de **Produto Escalar (Dot Product)** diretamente.
* Isso processa todos os números simultaneamente no hardware (vetorização), sendo ordens de grandeza mais rápido e deixando o código muito mais limpo.


### ☔ Estudo de Caso: Sistema de Decisão Pluviométrica
Vamos implementar um neurônio classificador binário que determina a ação: **"Levar guarda-chuva?"**.

#### 1. Definição do Vetor de Entradas (Sensores)
Para o computador, a informação dos sensores deve ser convertida em números e organizada em estruturas que chamamos de **Vetores** (ou no jargão de IA, *Tensores 1D*).

* $x_0$ **(Estado do Céu):** `1` (Nublado/Sim)
* $x_1$ **(Previsão Meteorológica):** `0.8` (80% de probabilidade de chuva)

#### 2. Definição do Vetor de Pesos (Conhecimento)
Os pesos representam a confiabilidade ou influência de cada sensor na decisão final.
* $w_0$ **(Peso Visual):** `2.0` (Baixa confiabilidade; nuvens enganam).
* $w_1$ **(Peso do App):** `10.0` (Alta confiabilidade; baseada em dados).

#### 🧮 Processamento Vetorial
O neurônio calculará o risco total através do produto escalar:
$$\text{Risco} = \vec{x} \cdot \vec{w} = (1 \times 2.0) + (0.8 \times 10.0) = 10.0$$

> 💡 **Cadê o Bias ($b$)?** > Na teoria acima, vimos o *bias* ($b$) sendo somado na equação. Neste primeiro exemplo prático, usaremos uma versão simplificada.

**Função de Ativação (Limiar):** Se o Risco calculado for superior a **5.0**, o neurônio "dispara" (ativação positiva: levar guarda-chuva).
