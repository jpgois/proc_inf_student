### 🧠 Cálculo da Mediana

A mediana é o valor que está exatamente no **meio** da lista. Mas atenção: o computador não sabe quem é o meio se os dados estiverem bagunçados. **A lista precisa estar ordenada** primeiro!

* **Se a lista tem tamanho ÍMPAR:** A mediana é o próprio elemento central.
* **Se a lista tem tamanho PAR:** Não existe um único "meio". A mediana será a média aritmética dos dois elementos centrais.

#### 📊 Visualizando o Processo

**1️⃣ Caso Ímpar (Ex: 5 elementos)**
Primeiro recebemos os dados desordenados e organizamos do menor para o maior. O elemento que sobrar isolado no centro é a nossa resposta.

```text
Vetor Desordenado: [ 9,  2, 15,  7,  4 ]

Vetor Ordenado:    [ 2,  4,  7,  9, 15 ]
                             ⬆
                         Mediana = 7
```
**2️⃣ Caso Par (Ex: 6 elementos)**
Aqui, após ordenar, não há um centro único. Pegamos os dois valores que dividem o vetor ao meio e calculamos a média entre eles.

```text
Vetor Desordenado: [ 20,  4,  9,  2, 15,  7 ]

Vetor Ordenado:    [  2,  4,  7,  9, 15, 20 ]
                              ⬆   ⬆
                              |   |
                    Mediana = (7 + 9) / 2 = 8                         
```

### Desmistificando a List Comprehension

- Se você olhou para a linha `[float(x) for x in entrada.split() if float(x) > 20]` e achou que o código foi escrito de trás para frente, **você não está sozinho!**

-A *List Comprehension* (Compreensão de Lista) comprime o código para que ele seja lido como uma frase direta (imitando o idioma inglês).

-Para entender, você só precisa dividir a linha em **3 blocos lógicos**, lendo da esquerda para a direita:


**1️⃣ O Resultado (O que eu quero na lista nova?)**
> `float(x)`
> *"Eu quero o valor de `x` convertido para decimal..."*

**2️⃣ A Fonte (De onde esse `x` está vindo?)**
> `for x in entrada.split()`
> *"...para cada pedacinho (`x`) da minha entrada de texto..."*

**3️⃣ O Filtro**
> `if float(x) > 20`
> *"...mas SÓ SE esse valor convertido for maior que 20."*

---

#### 🆚 Comparando os Mundos

        maiores.append(float(x))  # 1️⃣ O Resultado (ação no final)

## 🚀 Arrays de Verdade (`array` e `numpy`)

- Para processamento pesado (Engenharia, Data Science, Imagens), as Listas do Python são ineficientes.
- A solução é usar **Arrays Tipados**.

* **Característica:** Exigem que todos os dados sejam do **mesmo tipo** (só inteiros, ou só floats).
* **Vantagem:** São armazenados de forma contígua (lado a lado) na memória.

### 🛠️ Sintaxe Básica

Antes de testarmos o limite da memória, como criamos essas estruturas na prática?

A regra geral é simples: normalmente, nós pegamos uma lista comum do Python (ou geramos números) e "convertemos" para um Array tipado.

1. **`array`:** Exige que você passe um "código de tipo" (uma letra) para avisar o computador qual será o tamanho da caixa de memória. Exemplo: `'i'` para Inteiros, `'d'` para Decimais (Double).
2. **`numpy`:** É muito mais inteligente e amigável. Você passa o `dtype` (data type) ou, muitas vezes, ele mesmo adivinha o tipo automaticamente.

````python
import sys
import array
import numpy as np

def bytes_to_mb(bytes_val):
    return bytes_val / (1024 * 1024)

QTD = 1_000_000
lista_py = [0] * QTD
array_py = array.array('i', lista_py)
array_np = np.array(lista_py, dtype='int32')

print(f"--- Comparando Memória para {QTD} inteiros ---")

mem_lista = sys.getsizeof(lista_py)
mem_array = sys.getsizeof(array_py)
mem_numpy = array_np.nbytes

print(f"Python List: {bytes_to_mb(mem_lista):.2f} MB")
print(f"Array Nativo: {bytes_to_mb(mem_array):.2f} MB")
print(f"Numpy Array:  {bytes_to_mb(mem_numpy):.2f} MB")

print("\nConclusão: Estruturas do tipo Array (nativo ou NumPy) são muito mais eficientes em memória do que as Listas nativas do Python.")
````

### ⏱️ Medindo Performance (A Disputa Justa)

Além de economizar memória, os Arrays do NumPy são projetados para **velocidade**.
Mas para sermos justos, não podemos comparar o NumPy apenas com um `for` loop manual (que é naturalmente lento). O Python possui sua própria função otimizada de soma, o `sum()`.

Vamos usar o módulo `time` para cronometrar e fazer uma disputa tripla somando **10 milhões de números**:

1. 🐢 **Loop Manual:** O jeito mais lento, pois o Python lê linha por linha.
2. 🐕 **Função `sum()` nativa:** O jeito inteligente do Python. Ela é escrita na linguagem C (muito rápida), mas ainda precisa lidar com o peso das Listas flexíveis.
3. 🚀 **Função `np.sum()` do NumPy:** O topo da performance. Escrita em C e otimizada matematicamente para somar blocos inteiros de memória de uma só vez (Vetorização).

````python
import time
import numpy as np

QTD = 10_000_000
lista_grande = list(range(QTD))
array_grande = np.array(lista_grande, dtype='int32')

print(f"🏁 Disputa: Somando {QTD} números...\n")

# --- Competidor 1: Loop Manual ---
inicio = time.time()
soma = 0
for x in lista_grande:
    soma += x
fim = time.time()
print(f"🐢 1. Loop Manual (Python puro):  {fim - inicio:.4f} segundos")

# --- Competidor 2: Função sum() do Python ---
inicio = time.time()
soma = sum(lista_grande)
fim = time.time()
print(f"🐕 2. Função sum() nativa (Lista): {fim - inicio:.4f} segundos")

# --- Competidor 3: np.sum() do NumPy ---
inicio = time.time()
soma = np.sum(array_grande)
fim = time.time()
print(f"🚀 3. Função np.sum() (NumPy):     {fim - inicio:.4f} segundos")

print("-" * 45)
print("💡 Conclusão: O sum() nativo é excelente e esmaga o loop manual.")
print("Mas o NumPy ainda vence por lidar com Arrays puros e usar vetorização no processador!")
````

Um neurônio artificial opera realizando uma **combinação linear** de suas entradas. Antes de olharmos para a rede inteira acima, vamos entender a anatomia de um **único neurônio**:

#### Diagrama de um Único Neurônio Artificial

```text
   Entradas (x)      Pesos (w)        Processamento (Neurônio)
   ------------      ---------        -----------------------

   ( x0 ) ---------> [ w0 ] -----\
                                  \       Soma Ponderada
   ( x1 ) ---------> [ w1 ] ----------> Σ(xi * wi) + b ---> Saída
                                  /
   ( .. ) ---------> [ .. ] -----/
                                /
   ( bias ) -------------------/
     (b)
```


Estruturalmente, um neurônio é composto por:
1.  **Entradas ($x$):** Vetor de dados provenientes de sensores ou de outros neurônios.
2.  **Pesos ($w$):** Vetor de coeficientes que ponderam a importância de cada entrada (o "conhecimento" da rede).
3.  **Processamento (Soma Ponderada):** Ocorre o somatório do produto de cada entrada pelo seu peso correspondente.

Matematicamente (e simplificando), definimos o **Potencial de Ativação** ($u$) como:
$$u = \sum_{i=0}^{n} (x_i \cdot w_i) + b$$

### 🚀 Por que usar Arrays (NumPy) e não Listas?
* Se tentarmos multiplicar duas listas comuns (`vetor_x * vetor_w`), o Python retornará um **erro**. Listas nativas não sabem fazer diretamente operações matemáticas entre si.
* Em Python puro, a única saída seria usar um laço `for` para iterar, parear e multiplicar item a item, o que é computacionalmente lento para milhões de dados.
* Ao utilizarmos **Arrays** da biblioteca `numpy`, habilitamos a matemática vetorial. Podemos realizar a operação de **Produto Escalar (Dot Product)** diretamente.
* Isso processa todos os números simultaneamente no hardware (vetorização), sendo ordens de grandeza mais rápido e deixando o código muito mais limpo.


### ☔ Estudo de Caso: Sistema de Decisão Pluviométrica
Vamos implementar um neurônio classificador binário que determina a ação: **"Levar guarda-chuva?"**.

#### 1. Definição do Vetor de Entradas (Sensores)
Para o computador, a informação dos sensores deve ser convertida em números e organizada em estruturas que chamamos de **Vetores** (ou no jargão de IA, *Tensores 1D*).

* $x_0$ **(Estado do Céu):** `1` (Nublado/Sim)
* $x_1$ **(Previsão Meteorológica):** `0.8` (80% de probabilidade de chuva)

#### 2. Definição do Vetor de Pesos (Conhecimento)
Os pesos representam a confiabilidade ou influência de cada sensor na decisão final.
* $w_0$ **(Peso Visual):** `2.0` (Baixa confiabilidade; nuvens enganam).
* $w_1$ **(Peso do App):** `10.0` (Alta confiabilidade; baseada em dados).

#### 🧮 Processamento Vetorial
O neurônio calculará o risco total através do produto escalar:
$$\text{Risco} = \vec{x} \cdot \vec{w} = (1 \times 2.0) + (0.8 \times 10.0) = 10.0$$

> 💡 **Cadê o Bias ($b$)?** > Na teoria acima, vimos o *bias* ($b$) sendo somado na equação. Neste primeiro exemplo prático, usaremos uma versão simplificada.

**Função de Ativação (Limiar):** Se o Risco calculado for superior a **5.0**, o neurônio "dispara" (ativação positiva: levar guarda-chuva).
