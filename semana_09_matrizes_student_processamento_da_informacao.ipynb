### Matriz Diagonal (Versão Pythonica/Enxuta)

Podemos usar a função `all()` do Python.
* `all(iteravel)` retorna `True` apenas se **todos** os elementos do iterável forem verdadeiros.

A lógica abaixo lê-se: *"Retorne Verdadeiro se M[i][j] for zero para todo i, j onde i é diferente de j"*.

```python
def eh_diagonal_enxuta(M: list) -> bool:
    nro_linhas = len(M)
    nro_colunas = len(M[0])

    if nro_linhas != nro_colunas:
        return False

    return all(M[i][j] == 0
               for i in range(nro_linhas)
               for j in range(nro_colunas)
               if i != j)

print(eh_diagonal_enxuta([[1, 0], [0, 4]]))
print(eh_diagonal_enxuta([[1, 5], [0, 4]]))
```

## 5.4 Percorrendo a Diagonal Secundária

A lógica para encontrar o índice da coluna $j$ baseia-se na distância do fim da linha. Em uma matriz com $n$ colunas, a relação é:
$$
j = (n - 1) - i
$$

- Para $i=0 \implies j = (3-1) - 0 = 2 \implies (0, 2)$

## Multiplicação de Matrizes ($A \times B$)

A multiplicação matricial é um pouco mais complexa que a soma. Para multiplicar uma matriz $A$ de dimensão $m \times p$ por uma matriz $B$ de dimensão $p \times n$:

1.  O número de **colunas de A** deve ser igual ao número de **linhas de B** ($p$).
2.  A matriz resultante $C$ terá dimensão **$m \times n$**.

O elemento $C_{ij}$ é obtido pelo **produto escalar** da linha $i$ de $A$ pela coluna $j$ de $B$:

$$
C_{ij} = \sum_{k=1}^{p} A_{ik} \times B_{kj}
$$

**Exemplo 1: Produto Escalar (Vetores)**
A base da multiplicação é isso aqui:
$$
\begin{bmatrix}
1 & 2 & 3
\end{bmatrix}
\times
\begin{bmatrix}
4 \\ 5 \\ 6
\end{bmatrix}
=
[1\cdot4 + 2\cdot5 + 3\cdot6] = [32]
$$

**Exemplo 2: Caso Geral (Matrizes Retangulares)**
Vamos multiplicar uma matriz $A$ ($2 \times 3$) por uma matriz $B$ ($3 \times 2$). O resultado será uma matriz $2 \times 2$.

$$
A = \begin{bmatrix}
1 & 1 & 1 \\
2 & 2 & 2
\end{bmatrix}
\quad
B = \begin{bmatrix}
1 & 2 \\
3 & 4 \\
5 & 6
\end{bmatrix}
$$

Cálculo de cada elemento de $C$:
* $C_{00}$ (Linha 0 de A $\cdot$ Coluna 0 de B): $1\cdot1 + 1\cdot3 + 1\cdot5 = 9$
* $C_{01}$ (Linha 0 de A $\cdot$ Coluna 1 de B): $1\cdot2 + 1\cdot4 + 1\cdot6 = 12$
* $C_{10}$ (Linha 1 de A $\cdot$ Coluna 0 de B): $2\cdot1 + 2\cdot3 + 2\cdot5 = 18$
* $C_{11}$ (Linha 1 de A $\cdot$ Coluna 1 de B): $2\cdot2 + 2\cdot4 + 2\cdot6 = 24$

**Resultado:**
$$
C =
\begin{bmatrix}
9 & 12 \\
18 & 24
\end{bmatrix}
$$

###  Multiplicação Matricial (Versão "Enxuta")

Aqui a lógica fica densa. Usamos a função `sum()` para calcular o produto escalar (o laço `k`) dentro da construção da lista.

```python
def multiplica_matrizes_enxuta(A, B):
    if len(A[0]) != len(B):
        return "Matrizes com dimensões distintas"

    return [[sum(A[i][k] * B[k][j] for k in range(len(B)))
             for j in range(len(B[0]))]
            for i in range(len(A))]

C_enxuta = multiplica_matrizes_enxuta(A, B)
print("Produto Enxuto:")
for linha in C_enxuta:
    print(linha)
```

## Normalização e Ortogonalidade

**Normalização:** Um vetor unitário (norma = 1).
**Ortogonalidade:** Dois vetores são ortogonais se seu produto escalar for zero.

> **Nota Importante:** Assim como na normalização, ao verificar ortogonalidade com números de ponto flutuante, evitamos usar `== 0` devido a pequenos erros de precisão numérica. Usamos `np.isclose`.

```python
vetor = np.array([1 / np.sqrt(2), 1 / np.sqrt(2)])

norma = np.linalg.norm(vetor)

print(f"Norma calculada: {norma}")
print("É unitário?", np.isclose(norma, 1))

v1 = np.array([1, 0])
v2 = np.array([0, 1])

prod = np.dot(v1, v2)
print(f"\nProduto escalar: {prod}")

print("São ortogonais?", np.isclose(prod, 0))
```

```python
# 1. Verificação de Singularidade (Determinante = 0)
matriz = np.array([[1, 2], [2, 4]]) # Linhas são linearmente dependentes
det = np.linalg.det(matriz)
print(f"Determinante: {det}")
print("É singular?", np.isclose(det, 0))

# 2. Transposta e Simetria
# Matriz Simétrica: A == A.T
matriz_sim = np.array([[1, 2], [2, 1]])
eh_simetrica = np.array_equal(matriz_sim, matriz_sim.T)
print(f"\nMatriz:\n{matriz_sim}")
print("É simétrica?", eh_simetrica)

# 3. Verificação se é Quadrada
# shape[0] são linhas, shape[1] são colunas
matriz_ret = np.array([[1, 2], [3, 4], [5, 6]])
eh_quadrada = matriz_ret.shape[0] == matriz_ret.shape[1]
print(f"\nMatriz {matriz_ret.shape}:")
print("É quadrada?", eh_quadrada)
```

###  Soma e Multiplicação Matricial

No NumPy:
* `+` faz a soma elemento a elemento.
* `np.dot()` (ou o operador `@`) faz a multiplicação matricial real.
* `*` faz a multiplicação elemento a elemento (Cuidado!).

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])


soma = A + B
print("Soma (A + B):\n", soma)

# Multiplicação Matricial (Linha x Coluna)
# Pode usar np.dot(A, B) ou A @ B
produto = np.dot(A, B)
print("\nProduto Matricial (A . B):\n", produto)
```

##  Resolução de Sistemas Lineares

Dado o sistema $Ax = b$, queremos encontrar o vetor $x$.
O NumPy possui um solver otimizado em `numpy.linalg.solve`.

**Exemplo:**
$$
3x + y = 9 \\
x + 2y = 8
$$

Matricialmente:
$$
\begin{bmatrix} 3 & 1 \\ 1 & 2 \end{bmatrix} \cdot \begin{bmatrix} x \\ y \end{bmatrix} = \begin{bmatrix} 9 \\ 8 \end{bmatrix}
$$

```python
A = np.array([
    [3, 1],
    [1, 2]
])

b = np.array([9, 8])

x = np.linalg.solve(A, b)
print("Solução encontrada [x, y]:", x)

print("Verificação (A @ x):", np.dot(A, x))
```
