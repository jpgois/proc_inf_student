```python

### 🎮Exemplo: Criando um Menu

- No próximo exemplo , usaremos uma variável de texto (`opcao`) para controlar o laço.
- **Dica:**  usamos `.upper()` que garante que, mesmo que o usuário digite "s" minúsculo, o programa entenda como "S" maiúsculo e encerre o laço.

````python
# ESCREVA SEU CÓDIGO AQUI
raise NotImplementedError
````

````python
#um exemplo mais bonito de Menu...
from IPython.display import clear_output
import time

def menu_jogo():
    opcao = ""

    while opcao != "S":
        clear_output(wait=True)

        print("┌──────────────────────────┐")
        print("│    BEM-VINDO AO JOGO     │")
        print("└──────────────────────────┘")
        print(" [1] Iniciar Partida")
        print(" [2] Configurações")
        print(" [S] Sair")
        print("\n")

        time.sleep(0.5)

        opcao = input("Escolha uma opção: ").upper()

        if opcao != "S":
            print(f"Você escolheu: {opcao} (Processando...)")
            time.sleep(1)

    clear_output(wait=True)
    print("Jogo encerrado. Até logo!")

menu_jogo()
````

### 🕵️‍♀️ O Mistério do Zero

Por que deu zero? Vamos fazer o teste de mesa da primeira volta do laço:
1. `total` começa valendo `0`.
2. O código faz: `total = total * 1`.
3. Quanto é `0 * 1`? É **0**.
4. O `total` continua sendo 0 para sempre, não importa por quanto a gente multiplique!

**A Lição Matemática:**
* Na **Soma**, o elemento neutro é o **0** (porque $x + 0 = x$).
* Na **Multiplicação**, o elemento neutro é o **1** (porque $x \times 1 = x$).

Para consertar o fatorial, o acumulador DEVE nascer valendo 1.

### 🧠 Curiosidade Matemática (Pirâmides)

Essa soma é conhecida como "Números Piramidais".
Assim como na soma simples, existe uma fórmula direta para isso. Visualmente, estamos empilhando quadrados de tamanho $k \times k$.

**A Fórmula Mágica:**
$$S_n = \sum_{k=1}^{n} k^2 = \frac{n(n+1)(2n+1)}{6}$$

---

### 📐 Visualização (Prova Geométrica)
Se somarmos as camadas de uma pirâmide de base quadrada, temos :

| Camada ($k$) | Visual do Quadrado ($k^2$) | Área | Soma Acumulada |
| :---: | :---: | :---: | :---: |
| **Topo**<br>($k=1$) | $\color{cyan}{\blacksquare}$ | $1$ | **1** |
| **Meio**<br>($k=2$) | $\begin{matrix} \color{orange}{\blacksquare} & \color{orange}{\blacksquare} \\ \color{orange}{\blacksquare} & \color{orange}{\blacksquare} \end{matrix}$ | $4$ | $1 + 4 =$ **5** |
| **Base**<br>($k=3$) | $\begin{matrix} \color{green}{\blacksquare} & \color{green}{\blacksquare} & \color{green}{\blacksquare} \\ \color{green}{\blacksquare} & \color{green}{\blacksquare} & \color{green}{\blacksquare} \\ \color{green}{\blacksquare} & \color{green}{\blacksquare} & \color{green}{\blacksquare} \end{matrix}$ | $9$ | $5 + 9 =$ **14** |
